```python
import argparse
import os
from moviepy.editor import *

from IPython.display import Video


def edit_video(intro,outro,bg_music,video_path,bg_audio_track_level=0.10,padding = 1.5):
    

    # load video
    my_clip_ = VideoFileClip(video_path).fx(afx.audio_normalize)
    my_clip = my_clip_.subclip(0,my_clip_.duration)


    # pub bg music
    background_music = AudioFileClip(args_bg_music).fx(afx.audio_normalize)

    intro_audio_reduced = background_music.volumex(bg_audio_track_level).audio_loop(duration=my_clip.duration)
    new_myaudio = CompositeAudioClip([my_clip.audio, intro_audio_reduced])
    my_clip.audio = new_myaudio
    final_audio = my_clip.audio
    
    video_clips = [intro, my_clip, outro]  #  clips to crossfade

    video_fx_list = [video_clips[0]]
    # set padding to initial video
    idx = video_clips[0].duration - padding
    
    for video in video_clips[1:]:
        video_fx_list.append(video.set_start(idx).crossfadein(padding))
        idx += video.duration - padding

    final_video = CompositeVideoClip(video_fx_list)

    final_video  = final_video.set_duration(final_video.duration) 

    aud = final_video.audio.set_fps(44100)
    final_video = final_video.without_audio().set_audio(aud)
#     final_video.preview()
#     final_video.set_fps(25)
    final_video = final_video.fx(afx.audio_normalize)
    
    return final_video, final_audio


    
```

```python
args_intro = "/Volumes/GoogleDrive/My Drive/dev-mkt/021-marketing-videos/MV#01 - 101 Questions of AI/assets/AI101-intro.mp4" 
args_outro = "/Volumes/GoogleDrive/My Drive/dev-mkt/021-marketing-videos/MV#01 - 101 Questions of AI/assets/AI101-outro.mp4" 
args_input_dir =  "/Volumes/GoogleDrive/My Drive/dev-mkt/021-marketing-videos/MV#01 - 101 Questions of AI/test/"  
args_output_dir = "/Volumes/GoogleDrive/My Drive/dev-mkt/021-marketing-videos/MV#01 - 101 Questions of AI/test/edited/"
args_bg_music = "/Volumes/GoogleDrive/My Drive/dev-mkt/021-marketing-videos/MV#01 - 101 Questions of AI/assets/ES_Challenges Ahead - From Now On.mp3"

# create output dir (if it exists already we keep its content)
os.makedirs(args_output_dir, exist_ok=True)

# load intro an outro
# we subclip them since they have black screens that we don't want to load
intro_video = VideoFileClip(args_intro)
outro_video = VideoFileClip(args_outro)

# print (intro_video.duration)
# print (outro_video.duration)

intro = intro_video.subclip(0,intro_video.duration)
# .without_audio() 

outro = outro_video.subclip(0,outro_video.duration)
# outro = intro.audio_fadein(2)

# keep .mp4 files only
fn_list = sorted([fn for fn in os.listdir(args_input_dir) if fn.endswith(".mp4")])

# iterate video editing over all video in the input directory
for i, fn in enumerate(fn_list):

    video_path = os.path.join(args_input_dir, fn)


    edited_video, original_audio = edit_video(
        intro=intro,
        outro=outro,
        bg_music= args_bg_music,
        video_path=video_path
    )

    
    file_video = os.path.join(args_output_dir, fn.split(".")[0] + "-edited.mp4")
    print("INFO: WRITING FILE  {} ".format(file_video))
    
    edited_video.write_videofile(file_video, fps=24)
    
    
    original_audio.write_audiofile(os.path.join(args_output_dir, fn.split(".")[0] + "-edited.mp3"),fps = 44100)
    print("INFO: VIDEO {} / {} DONE".format(i + 1, len(fn_list)))
    print("-" * 50)
    print("-" * 50)
    print()
```
